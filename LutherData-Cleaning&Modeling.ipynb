**Consolidating and Encoding Categorical Variables**

```python
#map single family to 1 and others (condoor none to 0)
def map_fuc_type(x):
    if x in ['Single Family Home','single_family']:
        return 1
    return 0
```

```python
#map single family to 1 and others (condoor none to 0)
def map_fuc_cool(x):
    if x in  [' Central A/C', ' Central A/C, Ceiling fan(s)',' Window A/C Unit']:
        return 1
    return 0
```

```python
def map_fuc_view(x):
    if x in  ['View: Yes', 'View: Territorial', 'View: Bay', 'View: Partial','View: Sound']:
        return 1
    return 0
```

```python
def map_fuc_fireplace(x):
    if x in  ['Fireplace(s): Yes', 'Fireplace in Mstr BR', 'Fireplaces','Fireplace: Yes']:
        return 1
    return 0
```

```python
def school_rating_int(x):
    if x in ['NR']:
        return 0
    return int(x)
```

```python
from sklearn.metrics import r2_score
res_cv = {}
res_train = {}
res_cvr2={}
res_trainr2={}
for degree in range(5):
    
    # transforms the data to polynomials and scales data 
    est = make_pipeline(PolynomialFeatures(degree,interaction_only=True), StandardScaler(), LinearRegression())
    # calculates the MSE for CV
    scores = cross_val_score(est, X_train, y_train, cv=10, scoring='mean_squared_error') # cross val
    scoresr2 = cross_val_score(est, X_train,y_train, cv=10, scoring='r2')
    res_cv[degree] = np.mean(-scores)
    res_cvr2[degree]=np.mean(scoresr2)
    
    # calculates the MSE  and R2 for Training Set
    est.fit(X_train,y_train)
    y_pred = est.predict(X_train)
    res_train[degree] = np.sum((y_pred-y_train)**2)/len(y_train)
    res_trainr2[degree] = r2_score(y_test, est.predict(X_test))
    
```

```python

degrees = list(res_cv.keys())
mse_cv = list(res_cv.values())
mse_train = list(res_train.values())
r2_cv = list(res_cvr2.values())
r2_train = list(res_trainr2.values())

plt.figure(figsize=[15,5])
plt.subplot(1,2,1)
plt.plot(degrees,r2_cv)
plt.plot(degrees,r2_train)
plt.xlabel('Degree')
plt.ylabel('R2')
plt.title('log Scale')
plt.legend(['CV','Train'])
plt.axis([0, 5, 0, 1])

plt.subplot(1,2,2)
plt.semilogy(degrees,mse_cv)
plt.semilogy(degrees,mse_train)
plt.xlabel('Degree')
plt.ylabel('MSE')
plt.title('Log Scale')
plt.legend(['CV','Train'])

loc = np.argmin(list(res_cv.values()))
loc1=np.argmax(list(res_cvr2.values()))
print('The degree',degrees[loc],'had an MSE of:',mse_cv[loc])
print('The degree',degrees[loc1],'had an r2 of:',r2_cv[loc])
```

**Rigge and Lasso Regularization**

```python
from sklearn.linear_model import Ridge
import warnings
warnings.filterwarnings("ignore")

from sklearn.metrics import mean_squared_error

# Set up a figure and axes for 8 plots, 2 per row for 4 rows
fig, ax_rows = plt.subplots(4, 1, figsize=(6, 10))

# A helper function to plot the absolute value of the coefficients on the right-hand column plot
def plot_coefficients(est, ax, label=None, yscale='log'):
    coef = est.steps[-1][1].coef_.ravel() # flattens the data
    if yscale == 'log':
        ax.semilogy(np.abs(coef), marker='o', label=label)
        ax.set_ylim((1e-4, 1e10))
    else:
        ax.plot(np.abs(coef), marker='o', label=label)
        ax.set_ylabel('abs(coefficient)')
        ax.set_xlabel('coefficients')
        ax.set_xlim((1, 9))
        ax.set_ylim(-1000,1e6)

# Try out 4 different values of the RidgeRegression parameter alpha and watch how the resulting models change
# With higher values of alpha, more complex (more wiggly) models will be more punished and thus less likely
degree = 2
alphas = [1e-14, 1e-8, 1e-2, 100]
sse_tr_Ridge = []
sse_te_Ridge = []
r2_tr_Ridge = []
r2_te_Ridge = []
for alpha, ax_row in zip(alphas, ax_rows):
    ax_right = ax_row
    est = make_pipeline(PolynomialFeatures(degree,interaction_only=True), StandardScaler(), Ridge(alpha=alpha))
    est.fit(X_train, y_train)
    sse_tr_Ridge.append(mean_squared_error(y_train, est.predict(X_train)))
    sse_te_Ridge.append(mean_squared_error(y_test, est.predict(X_test)))
    r2_tr_Ridge.append(est.score(X_train,y_train))
    r2_te_Ridge.append(est.score(X_test,y_test))
    plot_coefficients(est, ax_right, label='Ridge(alpha=%r) coefficients' % alpha)
plt.tight_layout()
```

```python
plt.semilogx(alphas,sse_tr_Ridge,label='train',c='r')
plt.semilogx(alphas,sse_te_Ridge,label='test',c='b')
plt.ylim(0,0.5)
plt.legend()
plt.xlabel('Alpha')
plt.ylabel('SSE');
```

```python
plt.semilogx(alphas,r2_tr_Ridge,label='train',c='r')
plt.semilogx(alphas,r2_te_Ridge,label='test',c='b')
plt.ylim(0,1)
plt.legend()
plt.xlabel('Alpha')
plt.ylabel('Rˆ2');
```
